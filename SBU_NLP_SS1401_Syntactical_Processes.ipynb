#Working with Stanza (Persian)

```python
!pip install stanza
```

```python
import stanza

#POS tagging
input_sent = 'این جمله فقط برای آزمون است.'
#list of preprocesses: https://stanfordnlp.github.io/stanza/pipeline.html#processors
stanza.download('fa', verbose=True)
nlp = stanza.Pipeline('fa', processors='tokenize, lemma, pos, depparse',  download_method=None) #constituency not implemented in persian
doc = nlp(input_sent)
for sentence in doc.sentences:
    print('Showing POS tags of processed sentence: ')
    for word in sentence.words:
        print(f'word text: {word.text}, word pos: {word.pos}')
    print('-'*10)

#an abstract print of dependency tree
# for sentence in doc.sentences:
#     print(f'dependency tree: {sentence.dependencies}')


# a CoNLL print of dependency tree
for sent in doc.sentences:
   print('Showing dependency parsed result of processed sentence in CoNLL format: ')
   for word in sent.words:
     print(f'id: {word.id}\tword: {word.text}\thead id: {word.head}\thead: {sent.words[word.head-1].text if word.head > 0 else "root"}\tdeprel: {word.deprel}')
   print('-'*10)
```

```python
import stanza
stanza.download('en', verbose=True)
nlp = stanza.Pipeline(lang='en', processors='tokenize,pos,constituency', download_method=None)
doc = nlp('I saw the Eiffel tower flying to Paris')
for sentence in doc.sentences:
    print(sentence.constituency)
```

```python
from parsivar import Tokenizer, POSTagger

#POS tagging
input_sent = 'ما بسیار کتاب می‌خوانیم'
my_tokenizer = Tokenizer()
my_tagger = POSTagger(tagging_model="wapiti")  # tagging_model = "wapiti" or "stanford". "wapiti" is faster than "stanford"
tokens = my_tokenizer.tokenize_words(input_sent) 
text_tags = my_tagger.parse(tokens)
print(text_tags)
```

```python
from parsivar import DependencyParser, Tokenizer
from nltk.parse import DependencyGraph
from nltk.draw.tree import TreeView

#Dependency parsing
myparser = DependencyParser()
my_tokenizer = Tokenizer()
sents = "ما بسیار کتاب می‌خوانیم" 
sent_list = my_tokenizer.tokenize_sentences(sents)
# print(sent_list)
parsed_sents = myparser.parse_sents(sent_list)
for depgraph in parsed_sents: #depgraph is an instance of nltk.parse.DependencyGraph
  #use this site to show the graphical view of the tree (paste the conll result there): https://www.roshan-ai.ir/dependency-parse-tree/
  print(depgraph.to_conll(10))
```
